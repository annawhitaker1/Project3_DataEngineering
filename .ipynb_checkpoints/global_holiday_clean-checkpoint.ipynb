```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
monthly_passengers_df = pd.read_csv('monthly_passengers.csv')
holidays_travel_df = pd.read_csv('global_holidays.csv')
```

```python
monthly_passengers_df.head()
```

```python
holidays_travel_df.head()
```

```python
monthly_passengers_df.isnull().sum()
```

```python
monthly_passengers_df.dropna(inplace=True)
```

```python
monthly_passengers_df.duplicated().sum()
```

```python
monthly_passengers_df.info()
```

```python
holidays_travel_df.isnull().sum()
```

```python
holidays_travel_df.duplicated().sum()
```

```python
holidays_travel_df.drop_duplicates(inplace=True)
```

```python
holidays_travel_df.info()
```

```python
holidays_travel_df['Date'] = pd.to_datetime(holidays_travel_df['Date'])
```

```python
holidays_travel_df['Year'] = holidays_travel_df['Date'].dt.year
holidays_travel_df['Month'] = holidays_travel_df['Date'].dt.month
```

```python
merged_df = pd.merge(monthly_passengers_df, holidays_travel_df, on=['ISO3', 'Year', 'Month'], how='inner')
```

```python
merged_df.head(10)
```

```python
correlation = merged_df[[ 'Domestic', 'International']].corr()
print(correlation)
```

```python
# Aggregating by Year
agg_monthly = merged_df.groupby(['Year']).agg(
    Total=('Total', 'sum'),
    Domestic=('Domestic', 'sum'),
    International=('International', 'sum')
).reset_index()

pd.options.display.float_format = '{:,.2f}'.format

agg_monthly
```

```python
#Summary Stats of the Total, Domestic and International
summary_stats = merged_df[['Total', 'Domestic', 'International']].describe()
print(summary_stats)
```

```python
# Time series analysis
monthly_data = merged_df.groupby(['Year', 'Month'])['Total'].sum().reset_index()
monthly_data['Date'] = pd.to_datetime(monthly_data[['Year', 'Month']].assign(DAY=1))

plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_data, x='Date', y='Total', marker='o')
plt.title('Total Values Over Time')
plt.xlabel('Date')
plt.ylabel('Total')
plt.xticks(rotation=45)
plt.show()
```

```python
#Comparing data around holiday dates vs non-holiday dates.
holidays = merged_df[merged_df['Type'] == 'Public holiday']
non_holidays = merged_df[merged_df['Date'].isin(holidays['Date'])]

holiday_mean = holidays['Total'].mean()
non_holiday_mean = non_holidays['Total'].mean()

print(f"Mean Total for Holidays: {holiday_mean}")
print(f"Mean Total for Non-Holidays: {non_holiday_mean}")
```

```python
holidays_travel_df["Date"] = pd.to_datetime(holidays_travel_df["Date"])

holidays_travel_df["month_year"] = holidays_travel_df["Date"].dt.strftime("%B %Y")

holidays_travel_df = holidays_travel_df.rename(columns={"ADM_name": "country", "ISO3": "country_code"})

holidays_travel_mod_df = holidays_travel_df[["country_code","Date", "month_year", "Name","Type",]]

holidays_travel_mod_df.head()
```

```python
monthly_passengers_df["day"] = 1

monthly_passengers_df["month_year"] = pd.to_datetime(monthly_passengers_df[["Year", "Month","day"]]).dt.strftime("%B %Y")

monthly_passengers_df = monthly_passengers_df.rename(columns={"ISO3": "country_code","Total": "total_passengers_official",
                                                              "Total_OS": "total_passengers_open"})

monthly_passengers_mod_df = monthly_passengers_df[["country_code","Year", "Month", "day", "month_year", "total_passengers_official", "Domestic", "International", 
                                         "total_passengers_open"]]

monthly_passengers_mod_df.head()
```

```python
fk_country_df = holidays_travel_df[["country", "country_code"]]
fk_country_df = fk_country_df.drop_duplicates(subset=["country_code"])
fk_country_df
```

```python
monthly_passengers_mod_df.to_csv('modified_monthly_passengers.csv', index=False)
holidays_travel_mod_df.to_csv('modified_holidays_travel.csv', index=False)
fk_country_df.to_csv('new_country_code_table.csv', index=False)
```
